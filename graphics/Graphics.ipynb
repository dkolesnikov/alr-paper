```python
#Graphics
%matplotlib inline
import matplotlib.pyplot as plt
import prettyplotlib as ppl
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png')
import matplotlib as mpl
rc_cmufonts = {"font.family": "normal", 
                "font.serif": [], 
                "font.sans-serif": [], 
                "font.monospace": []}

fs = 12
font = {'size' : fs}
mpl.rc('font', **font)
mpl.rcParams.update(rc_cmufonts)
mpl.rcParams['text.usetex'] = True
%config InlineBackend.figure_format = 'svg'

#
name = "diffusion-4"
#name = "advection_diffusion"
#name = "laplace2d"

grid_list = [64, 128, 256]
font_size = 16

fig, ax = ppl.subplots(1, len(grid_list), figsize=(15, 5), sharey=True)

for i in xrange(len(grid_list)):
    name_grid = "%s_grid%d"%(name, grid_list[i])

    qq = loadmat('matlab/kpik_%s.mat'%name_grid)
    resnrm_kpik = qq['resnorm']
    qq = loadmat('matlab/rksm_%s.mat'%name_grid)
    resnrm_rksm = qq['resnorm']
    qq = loadmat('data/alr_%s.mat'%name_grid)
    resnrm_alr = qq['resnorm']
    #qq = loadmat('data/alr_mod_%s.mat'%name_grid)
    #resnrm_alr_mod = qq['resnorm']
    #minnorm = qq['minnorm']
 
    
    if i < len(grid_list) - 1:
        ax[i].semilogy(resnrm_kpik.T)#, color = 'b')
        ax[i].semilogy(resnrm_rksm.T)#, color = 'g')
        ax[i].semilogy(resnrm_alr.T)#, color = 'r')
        #ax[i].semilogy(minnorm.T)
        #ax[i].semilogy(resnrm_alr_mod.T)
    else:
        ax[i].semilogy(resnrm_kpik.T, label='KPIK')#, color = 'b')
        ax[i].semilogy(resnrm_rksm.T, label='RKSM')#, color = 'g')
        ax[i].semilogy(resnrm_alr.T, label='ALR')#, color = 'r')
        #ax[i].semilogy(minnorm.T, label = 'Minnorm')
        #ax[i].semilogy(resnrm_alr_mod.T, label='ALR mod')
    if i == 0:
        ax[i].set_ylabel('Residual', fontsize=font_size)
    if i == len(grid_list) / 2:     
        ax[i].set_xlabel('Number of solvers', fontsize=font_size)
    #ax[i].set_adjustable('box')    
    ax[i].set_title('Grid $%d\\times %d$' % (grid_list[i], grid_list[i]), fontsize=font_size)
    if i == len(grid_list) - 1:
        ax[i].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    ax[i].grid()

fig.tight_layout()
fig.savefig("data/%s.pgf"%name)    
print "data/%s.pgf"%name
```
